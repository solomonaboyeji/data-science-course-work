```python
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()


# Function to remove stop words and punctuation
def remove_stopwords_and_punctuation(text):   
    if type(text) != str:
        return text
        
    tokens = word_tokenize(text.lower())
    filtered_tokens = [lemmatizer.lemmatize(word) for word in tokens if word.lower() not in stop_words and word.lower() not in punctuation]
    return ' '.join(filtered_tokens)

# remove stop words and punctuations.
review_content_cleaned = cleaned_data_df['content'].apply(remove_stopwords_and_punctuation)
cleaned_data_df['cleaned_content'] = review_content_cleaned
```

```python
# -----------------------
# Word Frequency
# -----------------------
word_freq_cleaned = review_content_cleaned.str.split(expand=True).stack().value_counts()
# Plotting the top most frequent words
top_n_words = 50
plt.figure(figsize=(20, 12))
sns.barplot(x=word_freq_cleaned.head(top_n_words).values, y=word_freq_cleaned.head(top_n_words).index)
plt.title(f'Top {top_n_words} Most Frequent Words')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()
```

```python
from afinn import Afinn
# Initialize the AFINN sentiment analyzer
afinn = Afinn()

# -----------------------
# Sentiment Distribution
# -----------------------
def get_sentiment(text: str):
  # Get sentiment score for the word
  sentiment_score = afinn.score(text)
  sentiment_score = "NEGATIVE" if sentiment_score < 0 else "POSITIVE" if sentiment_score > 0 else "NEUTRAL"
  return sentiment_score

for index, text in enumerate(review_content_cleaned):
    if type(text) == str:
      sentiment = get_sentiment(text)
      cleaned_data_df.at[index, "sentiment"] = sentiment
      cleaned_data_df.at[index, "language"] = get_language(text)
    else:
      cleaned_data_df.at[index, "sentiment"] = "NOT_DETECTED"
```

```python
from datetime import datetime
# # Feature Engineering: Year of Date Written
def get_year_from_date(at_date_string: str):
  date_object = datetime.strptime(at_date_string, "%Y-%m-%d %H:%M:%S")
  return date_object.year

# Feature Engineering: Year/Month of Date Written
def get_year_and_month_from_date(at_date_string: str):
  date_object = datetime.strptime(at_date_string, "%Y-%m-%d %H:%M:%S")
  month_str = int(date_object.month)
  return f"{date_object.year}/{month_str}"

# Mapping months to seasons
seasons_dict = {
    1: "Winter", 2: "Winter", 3: "Spring",
    4: "Spring", 5: "Spring", 6: "Summer",
    7: "Summer", 8: "Summer", 9: "Fall",
    10: "Fall", 11: "Fall", 12: "Winter"
}

# Get the season for a given month
def get_season(year_month):
    month = year_month.split("/")[1]
    return seasons_dict[int(month)]

if "at_year_month" not in cleaned_data_df.columns.to_list():
  year_and_month_df = pd.DataFrame(cleaned_data_df["at"].apply(get_year_and_month_from_date))
  cleaned_data_df = cleaned_data_df.join(year_and_month_df, how="right", rsuffix="_year_month")

if "at_year" not in cleaned_data_df.columns.to_list():
  year_df = pd.DataFrame(cleaned_data_df["at"].apply(get_year_from_date))
  cleaned_data_df = cleaned_data_df.join(year_df, how="right", rsuffix="_year")

if "at_year_month_season" not in cleaned_data_df.columns.to_list():
  seasons_df = pd.DataFrame(cleaned_data_df['at_year_month'].apply(get_season))
  cleaned_data_df = cleaned_data_df.join(seasons_df, how="right", rsuffix="_season")
```

```python
# Word Embeddings
import gensim.downloader as api

# Load pre-trained Word2Vec model -> downloads close to 1.6gb of model
word2vec_model = api.load('word2vec-google-news-300')

def get_word_vectors(text):
    words = text.split()
    word_vectors = [word2vec_model[word] for word in words if word in word2vec_model.key_to_index]
    return np.mean(word_vectors, axis=0) if len(word_vectors) > 0 else np.zeros(300)

# Apply word embeddings to the review content
data['word_vectors'] = data['cleaned_content'].fillna('').apply(get_word_vectors)
```

```python
import numpy as np
from sklearn.manifold import TSNE

# Perform t-SNE on the word embeddings
tsne = TSNE(n_components=2, random_state=42)
tsne_embeddings = tsne.fit_transform(np.stack(data['word_vectors'].values))

# Create a scatterplot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=tsne_embeddings[:, 0], y=tsne_embeddings[:, 1], hue=data['category'], palette='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE Scatterplot of Reviews')
plt.show()
```
